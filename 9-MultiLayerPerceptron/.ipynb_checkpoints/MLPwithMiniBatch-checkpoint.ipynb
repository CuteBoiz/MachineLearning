# MLP & Back Propagation

**1. Feed forward**
$$ A^{(0)} = X $$

$$ Z^{(1)} = W^{(1)}.A^{(0)} + B^{(1)} $$

$$ A^{(1)} = max(Z^{(1)}, 0) $$

$$ Z^{(2)} = W^{(2)}.A^{(1)} + B^{(2)} $$

$$ \hat{Y} = A^{(2)} = \sigma (Z^{(2)}) $$

**2. Back Propagation**
$$ E^{(2)}=\triangledown_{Z^{(2)}}=\frac{1}{N}(A^{(2)}-Y) $$

$$ \triangledown_{W^{(2)}}=A^{(1)}.E^{(2)T};\ \triangledown_{b^{(2)}}=\sum^{N}_{n=1}e^{(2)}_{n} $$

$$ E^{(1)}=(W^{(2)}.E^{(2)})\odot f^{'}(Z^{(1)}) $$

$$ \triangledown_{W^{(1)}}=A^{(0)}.E^{(1)T} = X.E^{(1)T};\ \triangledown_{b^{(1)}}=\sum^{N}_{n=1}e^{(1)}_{n} $$


```python
import numpy as np
import matplotlib.pyplot as plt

def softmax_stable(Z):
    e_Z = np.exp(Z - np.max(Z, axis=1, keepdims=True))
    return e_Z/e_Z.sum(axis=1, keepdims=True)

def crossentropy_loss(Yhat, y):
    id0 = range(Yhat.shape[0])
    return -np.mean(np.log(Yhat[id0, y]))

def mlp_init(d0, d1, d2):
    W1 = 0.01*np.random.rand(d0, d1)
    b1 = np.zeros(d1)
    W2 = 0.01*np.random.rand(d1, d2)
    b2 = np.zeros(d2) 
    return (W1, b1, W2, b2)

def mlp_predict(X, W1, b1, W2, b2):
    Z1 = X.dot(W1) + b1
    A1 = np.maximum(Z1, 0)
    Z2 = A1.dot(W2) + b2
    return np.argmax(Z2, axis=1)

def mlp_fit(X, y, W1, b1, W2, b2, lr=0.1, n_epoches=10000, batch=10):
    loss_hist = []
    N = X.shape[0]
    for ep in range(n_epoches):      
        mix_ids = np.random.permutation(N)
        X_indices = X[mix_ids]
        y_indices = y[mix_ids]
        
        #Caculate loss
        if ep % 1000 == 0:
            Z1 = X.dot(W1) + b1
            A1 = np.maximum(Z1, 0) #ReLU
            Z2 = A1.dot(W2) + b2
            A2 = softmax_stable(Z2)
            loss = crossentropy_loss(A2, y)
            print("iter: %d, loss: %f" %(ep, loss))
            loss_hist.append(loss)
            
        for i in range(0, N, batch):
            X_i = X_indices[i:i+batch]
            y_i = y_indices[i:i+batch]
            
            #feed forward
            Z1 = X_i.dot(W1) + b1
            A1 = np.maximum(Z1, 0) #ReLU
            Z2 = A1.dot(W2) + b2
            A2 = softmax_stable(Z2)
            
            #back propagation
            id0 = range(A2.shape[0])
            A2[id0, y_i] -= 1 #A2 - Y
            E2 = 1/N*A2
            dW2 = A1.T.dot(E2)
            db2 = np.sum(E2, axis=0)
            E1 = E2.dot(W2.T)
            E1[Z1 <= 0] = 0 # gradient of ReLU
            dW1 = X_i.T.dot(E1)
            db1 = np.sum(E1, axis=0)
            
            #Update weights
            W1 += -lr*dW1
            b1 += -lr*db1
            W2 += -lr*dW2
            b2 += -lr*db2
            
    return (W1, b1, W2, b2, loss_hist)
```

## Data

```python
import pandas as pd

data_file = '../data/Iris.csv'
data = pd.read_csv(data_file)
inputs, outputs = data.iloc[:, 0:5], data.iloc[:, 5]
X_id, y = np.array(inputs.values), np.array(outputs.values)

X = X_id[:, 1:5]
y = np.asarray([0]*50 + [1]*50 +[2]*50)
y.reshape(-1, 1)
data
```

```python
d0 = 4
d1 = 500
d2 = 3

(W1, b1, W2, b2) = mlp_init(d0, d1, d2)
(W1, b1, W2, b2, loss) = mlp_fit(X, y, W1, b1, W2, b2, lr=0.1, n_epoches=20000)

y_pred = mlp_predict(X, W1, b1, W2, b2)
acc = 100*np.mean(y_pred == y)
print("acc: %.2f %%" %acc)

plt.plot(loss)
plt.show()
```
